### Asumptions for the check distance function
##### Checks if the smma distance is overcrowded or not


* The 200 smma is always greater than the 50 smma and 21 smma (and vice versa)
* The 50 smma is alays greater than the 21 smma (and vice versa)


```python
def check_dist(sma1,sma2,sma3,allowed_dist):
    diff_dist1=sma2-sma1
    diff_dist2=sma3-sma2
    return_value=True
    if diff_dist1<allowed_dist:
        return_value=False
    if diff_dist2<allowed_dist:
        return_value=False
        
    return return_value

    
```

```python
def check_engulf(n_candle,p_candle):
    return_vale=True
    f_bodysize=abs(n_candle['Close']-n_candle['Open'])
    p_bodysize=abs(p_candle['Close']-p_candle['Open'])
    bd_diff=f_bodysize-p_bodysize
    
    
    if candletype(n_candle)==candletype(p_candle):
        return_vale=False

    else:
       
        if bd_diff<p_bodysize:
            return_vale=False

      
            
    return return_vale            
            
            
    
    
```

```python
def confirn_dist_smma(candle,allowed_d):
    return_value=True
    if candletype(candle)==1:
        distance=abs(candle['smma_21']-candle['High'])
        if distance>allowed_d:
            return_value=False
    else:
        distance=abs(candle['smma_21']-candle['Low'])
        if distance>allowed_d:
            return_value=False
    
    return return_value

```

```python
def check_overunder(candle):
    return_value=True
    if candletype(candle)==1:
        if candle['Open']<candle['smma_21']:
            return_value=False
        if candle['Open']<candle['smma_50']:
            return_value=False
        if candle['Open']<candle['smma_200']:
            return_value=False
    else:
        if candle['Open']>candle['smma_21']:
            return_value=False
        if candle['Open']>candle['smma_50']:
            return_value=False
        if candle['Open']>candle['smma_200']:
            return_value=False
            
    return return_value

    
```

```python
def disjunction(data,index,tradetype):
    if tradetype==1:
        res_point=resistance(data,index,2,1)
        rsi_point=data.loc[res_point]['rsi']
        if rsi_point<40:
            return True
        else:
            return False
    else:
        sup_point=support(data,index,2,1)
        rsi_point=data.loc[sup_point]['rsi']
        if rsi_point>60:
            return True
        else:
            return False
        
    
    
    
```

```python
def convert_timestamps(data):
    new_data = []
    for entry in data:
        readable_start = datetime.datetime.utcfromtimestamp(entry[1]).strftime('%Y-%m-%d %H:%M:%S')
        readable_end = datetime.datetime.utcfromtimestamp(entry[2]).strftime('%Y-%m-%d %H:%M:%S')
        new_entry = (entry[0], readable_start, readable_end, entry[3], entry[4])
        new_data.append(new_entry)
    return new_data

```
